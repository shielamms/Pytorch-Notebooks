### Numpy Version

```python
import numpy as np
```

```python
# N = sample size
N = 64

# D_in = number of features
D_in = 1000

# H = Number of hidden layers
H = 100

# D_out = Number of output layers
D_out = 10
```

```python
x = np.random.randn(N, D_in)
y = np.random.randn(N, D_out)
```

```python
w1 = np.random.randn(D_in, H)
w2 = np.random.randn(H, D_out)
```

```python
learning_rate = 1e-6
```

```python
for t in range(500):
    # Forward pass
    z = x.dot(w1)
    z_relu = np.maximum(z, 0)
    y_pred = z_relu.dot(w2)
    
    # Compute the loss
    loss = np.square(y_pred - y).sum()
    print(t, loss)
    
    # Backprop
    grad_y_pred = 2.0 * (y_pred - y)
    grad_w2 = z_relu.T.dot(grad_y_pred)
    grad_z_relu = grad_y_pred.dot(w2.T)
    grad_z = grad_z_relu.copy()
    grad_z[z < 0] = 0
    grad_w1 = x.T.dot(grad_z)
    
    # Update weights
    w1 -= learning_rate * grad_w1
    w2 -= learning_rate * grad_w2
```

### PyTorch Version

```python
import torch
```

```python
dtype = torch.float
device = torch.device('cpu')
```

```python
# N = sample size
N = 64

# D_in = number of features
D_in = 1000

# H = Number of hidden layers
H = 100

# D_out = Number of output layers
D_out = 10
```

```python
x = torch.randn(N, D_in, device=device, dtype=dtype)
y = torch.randn(N, D_out, device=device, dtype=dtype)
```

```python
w1 = torch.randn(D_in, H, device=device, dtype=dtype)
w2 = torch.randn(H, D_out, device=device, dtype=dtype)
```

```python
learning_rate = 1e-6
```

```python
for t in range(500):
    # Forward pass
    z = x.mm(w1)
    z_relu = z.clamp(min=0)
    y_pred = z_relu.mm(w2)
    
    # Compute the loss
    loss = (y_pred - y).pow(2).sum().item()
    if t % 100 == 99:
        print(t, loss)
    
    # Backprop
    grad_y_pred = 2.0 * (y_pred - y)
    grad_w2 = z_relu.t().mm(grad_y_pred)
    grad_z_relu = grad_y_pred.mm(w2.T)
    grad_z = grad_z_relu.clone()
    grad_z[z < 0] = 0
    grad_w1 = x.t().mm(grad_z)
    
    # Update weights
    w1 -= learning_rate * grad_w1
    w2 -= learning_rate * grad_w2
```
